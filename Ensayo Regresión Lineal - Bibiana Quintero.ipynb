## <span style='font-family:Segoe UI'> **Introducción**


<span style='font-family:Segoe UI'> Uno de los más importantes campos en estadística es el estudio de regresión. Existen muchas técnicas de análisis de regresión, una de ellos es la regresión linear; es la técnica más simple usada en los métodos de regresión que busca las relaciones existentes entre diferentes variables. Una de sus ventajas es la fácil interpretación de los resultados.

    
<span style='font-family:Segoe UI'> Al implementar una regresión linear de alguna variable dependiente $y$ con un conjunto de variables independientes $x=(x_1,..., x_n)$, donde $n$ es el número de predictores, se asume una relación lineal entre $y$ y $x: y = \beta_0 + \beta_1 x_1 + ... + \beta_n x_n + \epsilon $ donde $\beta_0 , \beta_1 , ... , \beta_n$ son los coeficientes de la regresión y $\epsilon$ el error.
    
<span style='font-family:Segoe UI'> De acuerdo con lo anterior, el modelo de regresión se expresa en la siguiente ecuación:
    
$$Y = \beta_0 + \beta_1 X_1 $$

## <span style='font-family:Segoe UI'> **Breve historia**

<span style='font-family:Segoe UI'>Legendre en 1805, publicó la primera forma de regresión lineal denomiada como el método de los mínimos cuadrados ordinarios. Luego Gauss, desarrolló más profundamente dicho método, incluyendo una versión del teorema de Gauss-Márkov.
    
<span style='font-family:Segoe UI'>El término regresión se utilizó por primera vez en el estudio de variables antropométricas; se comparó la estatura de padres e hijos, obteniendo como resultado que los hijos cuyos padres tenían una estatura muy superior al valor medio, tendían a igualarse a este, mientras que aquellos cuyos padres eran muy bajos tendían a reducir su diferencia respecto a la estatura media. El término lineal se utiliza para diferenciar el tipo de regresión de las demás técnicas. Para más información: [Wikipedia](https://es.wikipedia.org/wiki/Regresi%C3%B3n_lineal#:~:text=La%20primera%20forma%20de%20regresi%C3%B3n,del%20teorema%20de%20Gauss%2DM%C3%A1rkov.)
 
<span style='font-family:Segoe UI'>Los modelos de regresión lineal tienen muchas aplicaciones en diversos sectores, como la predicción de ventas en el sector comercio, la predicción del precio de la vivienda en el sector inmobiliario, la predicción del tiempo en meteorología, la predicción del precio de las acciones en Finanzas, entre otros. 

## <span style='font-family:Segoe UI'> **Planteamiento del algoritmo - Ejemplo de regresión lineal**
    
<span style='font-family:Segoe UI'> **1. Importar librerías**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

<span style='font-family:Segoe UI'> **2. Importar conjunto de datos**

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

<span style='font-family:Segoe UI'> **3. Dividir el conjunto de datos en conjunto de entrenamieento y de prueba**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

<span style='font-family:Segoe UI'> **4. Entrenamiento del modelo de regresión lineal simple en el conjunto de entrenamiento**

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

<span style='font-family:Segoe UI'> **5. Predecir los resultados del conjunto de prueba**

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

<span style='font-family:Segoe UI'> **6. Visualización de los resultados del conjunto de entrenamiento**

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salario vs Experiencia (Training set)')
plt.xlabel('Años de experiencia')
plt.ylabel('Salario')
plt.show()
```

<span style='font-family:Segoe UI'>  **7. Visualización de los resultados del conjunto de prueba**

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salario vs Experiencia (Test set)')
plt.xlabel('Años de experiencia')
plt.ylabel('Salario')
plt.show()
```
